```python
#pygame_bricks
import pygame
import sys
import time
import random

from pygame.locals  import *

# 색상, 크기 정의 등 미리 해놓을 것들
BLACK = (0,0,0)
WHITE = (255,255,255)
RED = (255,0,0)


WINDOW_WIDTH = 600
WINDOW_HEIGHT = 400

pygame.init()
window=pygame.display.set_mode( (WINDOW_WIDTH, WINDOW_HEIGHT), 0, 32 )

# 처음에  공, 이동막대, 벽돌들
paddle = pygame.Rect(300, 350, 60, 10)   # 위치(600/2, 400보다 조금 작은 350), 크기(60,10) 가로막대
ball = pygame.Rect(300,250,10,10) # 위치 중앙 , 이동막대보다 조금 위에, 크기(10,10) 
ball_dx = 2
ball_dy = -2   # 방향지정

bricks = []
for i in range(5):
    for j in range(3) :
        bricks.append(pygame.Rect( 50 + 100 * i , 50+20*j, 60,10   ) )
        

pygame.key.set_repeat(1,20) # 키 누르고 있는 기능 제공 1,50,  1,40

while True :
    for event in pygame.event.get() :
        if event.type==QUIT :
            pygame.quit()
            sys.exit()
        # event 확인, ket event 
        elif event.type == KEYDOWN:
            if event.key == K_LEFT and paddle.left > 0:
                paddle.left -= 2
            elif event.key == K_RIGHT and paddle.right < WINDOW_WIDTH:
                paddle.right+=2 
    #반복해서 하고 싶은 동작
    # 공움직이기
    ball.left += ball_dx
    ball.top += ball_dy

    # 화면벽과 충돌
    if ball.left <=0 or ball.right >= WINDOW_WIDTH : # 양옆 충돌
        ball_dx*=-1
    if ball.top <= 0 or ball.top >= WINDOW_HEIGHT : # 위아래 충돌
        ball_dy*=-1
        
    # paddle (이동막대)랑 충돌
    if ball.colliderect(paddle) :
        ball_dy *= -1

    # bricks 랑 충돌
    hit_index = ball.collidelist(bricks)  #bricks 전부 검사해서 충돌한 게 있는지 없는지
    if hit_index != -1 :
        bricks.pop(hit_index)
        ball_dy *= -1
        
    # 바닥에 닿으면 종료
    if ball.top >= WINDOW_HEIGHT:
        font = pygame.font.Font("NanumGothic.ttf", 70)   # 폰트랑 글자 크기
        fail_surface = font.render("게임실패", 1, WHITE)  # 글자를 객체로
        fail_rect = fail_surface.get_rect()              # 객체의 사각형 객체
        fail_rect.center = (WINDOW_WIDTH/2, WINDOW_HEIGHT/2)  # 사각형 객체 위치 지정
        window.blit(fail_surface, fail_rect)
        pygame.display.flip()
        pygame.time.wait(3000)
        pygame.quit()
        sys.quit()
        print("게임종료")

    # bricks가 사라져도 종료
    if len(bricks) == 0:
        font = pygame.font.Font("NanumGothic.ttf", 70)
        success_surface = font.render("클리어", 1, WHITE)
        success_rect = success_surface.get_rect()
        success_rect.center = (WINDOW_WIDTH/2, WINDOW_HEIGHT/2)
        window.blit(success_surface, success_rect)
        pygame.display.flip()
        pygame.time.wait(3000)
        pygame.quit()
        sys.quit()
        print("클리어")
    
    
    # 그리기
    window.fill(BLACK)
    pygame.draw.rect( window, WHITE, paddle) # 이동막대가 하얗게
    pygame.draw.rect( window, WHITE, ball) 

    for i in bricks :
        pygame.draw.rect(window,RED, i)
                
    # 화면 업데이트
    pygame.display.flip()    # 모든 surface 업데이트
    # pygame.display.update()  # 매개변수가 없으면 flip이랑 비슷. 매개변수가 있으면 특정 surface만 
    pygame.time.delay(10)  # 0.01초

```
