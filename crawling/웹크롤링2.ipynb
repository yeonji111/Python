```python
#네이버 웹툰 데이터 수집
# 인기순, 업데이트순, 조회수, 별점순  클릭    반복문으로 해야되지않을까???
# 그럼 에러가 날 확률 99%
# orderby별  반복문
#     요일별 반복문
#         웹툰 제목 출력 쫘라락


from selenium import webdriver
import time
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import pandas as pd 

selctor= By.CSS_SELECTOR
driver=webdriver.Chrome()   # driver.  
url= "https://comic.naver.com/webtoon"
driver.get(url)
driver.implicitly_wait(2)  
div_tab=driver.find_element(selctor , "div[role='tablist']" )
button_list= div_tab.find_elements(selctor , "button")

for button in button_list :
    button.click()
    time.sleep(2)
    div_rect = driver.find_element(selctor , "div.WeekdayMainView__daily_all_wrap--UvRFc")
    div_week_list= div_rect.find_elements(selctor, "div.WeekdayMainView__daily_all_item--DnTAH")
    for div_day in div_week_list : 
        h3=div_day.find_element(selctor,"h3")  # '월요웹툰'
        li_list=div_day.find_elements(selctor, "ul.WeekdayMainView__daily_list--R52q0 li.DailyListItem__item--LP6_T")
        #ul밑에 li들   li하나는 웹툰하나.
        
        for li in li_list :
            print(button.text, h3.text   ,li.find_element(selctor, "span.text").text )
                                 










```
